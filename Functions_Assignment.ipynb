Q1> What is the difference between a function and a method in Python?

Ans> **Function**: A function is a block of code that performs a specific task and can be called to execute that task. Functions are independent and can be defined outside of any class.

**Method**: A method is a function that is associated with an object and is defined inside a class. Methods usually operate on the data that belongs to the object (instance of the class).

**Example**:



````python
# Function
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")

# Method inside a class
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, {self.name}!")

person = Person("Alice")
person.greet()  # This is a method call
````

Q5> What are iterators in Python and how do they differ from iterables?

Ans> **Iterable**: An iterable is an object that can be iterated (looped) over, like lists, tuples, and strings. It defines a __iter__() method.

**Iterator**: An iterator is an object that represents a stream of data and is used to iterate over an iterable. It defines the __next__() method.

Example:

````python
# Iterable example
my_list = [1, 2, 3]
for item in my_list:  # The list is iterable
    print(item)

# Iterator example
my_iterator = iter(my_list)  # Creates an iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
````

Q6> Explain the concept of generators in Python and how they are defined.

Ans> A generator is a function that returns an iterable set of items, one at a time, using the yield keyword. Generators are more memory efficient than normal functions as they yield items on demand.

Example:

````python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
for item in gen:
    print(item)  # Output: 1 2 3
````

Q7> What are the advantages of using generators over regular functions?

Ans> Advantages of Using Generators Over Regular Functions

1. **Memory Efficiency**:
   - Generators do not store the entire sequence in memory; they yield values one at a time. This is particularly useful when dealing with large datasets or streams of data that cannot fit entirely into memory.
   
2. **Lazy Evaluation**:
   - Generators compute values only when needed, which makes them more efficient when dealing with large or infinite sequences. They do not evaluate all the values upfront, unlike regular functions that return the entire result at once.

3. **Improved Performance**:
   - Because generators yield values one at a time, they often lead to improved performance, especially when you only need to process a small portion of a large dataset at a time.

4. **Cleaner Code**:
   - Generators allow you to write cleaner and more readable code for iterating through sequences, avoiding the need to manually manage iteration or state.

### Example: Using Generators vs Regular Functions

#### Regular Function:
A regular function that returns a list of squared numbers might look like this:
```python
def squares_regular(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

squares_list = squares_regular(5)
print(squares_list)  # Output: [0, 1, 4, 9, 16]
```
- This function stores all the squares in memory at once, which can be inefficient if `n` is large.

#### Generator Function:
The same logic implemented as a generator function:
```python
def squares_generator(n):
    for i in range(n):
        yield i * i

gen = squares_generator(5)
for square in gen:
    print(square)  # Output: 0 1 4 9 16
```
- This generator yields each squared number one at a time, without storing the entire sequence in memory.

### Comparison:
1. **Memory Usage**:
   - `squares_regular` creates a list to store all the squared numbers, consuming memory proportional to `n`.
   - `squares_generator` only computes one value at a time and doesnâ€™t store the entire sequence, making it memory-efficient.

2. **Lazy Evaluation**:
   - The generator yields values only when requested by the `for` loop, while the regular function computes all values upfront.

Thus, **generators** are more efficient when handling large datasets or when you only need to process the sequence lazily.

Q10> What is the difference between map(), reduce(), and filter() functions in Python?

Ans> map(): Applies a function to all items in an iterable and returns a map object.

reduce(): Applies a function cumulatively to the items of an iterable, reducing it to a single value.

filter(): Filters the items of an iterable by applying a function that returns True or False.

Example:

````python
from functools import reduce

# map example
numbers = [1, 2, 3]
print(list(map(lambda x: x*2, numbers)))  # Output: [2, 4, 6]

# filter example
print(list(filter(lambda x: x % 2 == 0, numbers)))  # Output: [2]

# reduce example
print(reduce(lambda x, y: x + y, numbers))  # Output: 6 (sum of the numbers)
````

Q1> Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

````python
def sum_even_numbers(numbers):
    # Initialize the sum to 0
    total = 0

    # Iterate over the list of numbers
    for number in numbers:
        # Check if the number is even
        if number % 2 == 0:
            total += number  # Add the even number to the sum

    return total

# Example usage
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = sum_even_numbers(numbers)
print(result)  # Output: 20 (2 + 4 + 6 + 8)
````

Q4> Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(number):
    # Numbers less than 2 are not prime
    if number < 2:
        return False

    # Check for divisibility from 2 to the square root of the number
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False

    return True

# Example usage: Checking prime numbers between 1 and 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
````

Q5> Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
class FibonacciIterator:
    def __init__(self, n):
        # Initialize the number of terms to generate
        self.n = n
        self.count = 0
        self.a, self.b = 0, 1  # Starting values for Fibonacci sequence

    def __iter__(self):
        # The __iter__ method returns the iterator object itself
        return self

    def __next__(self):
        # Stop iteration if the number of terms has been reached
        if self.count >= self.n:
            raise StopIteration

        # Return the current Fibonacci number
        current = self.a
        # Update the values of a and b for the next Fibonacci number
        self.a, self.b = self.b, self.a + self.b
        self.count += 1

        return current

# Example usage: Create a Fibonacci sequence iterator for the first 10 terms
fibonacci = FibonacciIterator(10)

# Iterate and print the Fibonacci numbers
for num in fibonacci:
    print(num)
````

Q6> Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
````

Q9> Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

````python
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 25, 30, 35, 40]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (9/5) * celsius + 32

# Use map() to apply the function to each element of the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the result
print(fahrenheit_temperatures)
````

Q10> Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'  # Define vowels (both lowercase and uppercase)
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Use filter() to keep only characters that are not vowels
result = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(result)
````
